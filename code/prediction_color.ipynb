```python
import cv2
import tensorflow as tf
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from keras.preprocessing.image import img_to_array
import numpy as np 
import matplotlib.pyplot as plt
import os
import PIL.Image as Image
```

```python
path = "..\projectimg\\CARS\\types\\test\\suv\\suv9.jpg" #example
def prediction(path):
    global image ,index
    image = Image.open(path).resize((270,180))
    image = (np.array(image)/255)
    plt.imshow(image)
    plt.show()
    print(image.shape)
    image = np.expand_dims(image,axis =0)
    print(image.shape)
    image = np.vstack([image])
    print(image.shape)
    model = tf.keras.saving.load_model("hf://shripadd/car-color-detection")
    prediction = model.predict(image)
    index  = np.argmax(prediction)
    print("The Result is : ",prediction)
prediction(path)
if index == 0:
    print("The Car Is : BLACK")
elif index == 1:
    print("The Car Is : BLUE")
elif index == 2 :
    print("The Car Is : RED")
elif index ==3 :
    print("The Car Is : WHITE")
```
